### <span style="color:lightgreen">**The `print` function**</span>
***

When we run a code cell, we usually get just one result, but sometimes we need to display more than one. <br>
To do this, we can use the `print` function, which allows us to display results in Python.

The `print` function can show any type of object, including variables that contain floats, integers, strings, etc.

````python
print("Hello world!!!")
````

***
**Example: Mine production**

A copper mine produced the following tons of concentrate over 4 weeks: 910, 825, 1070, 940.

We can calculate the total monthly production by adding these 4 amounts and assigning the result to the variable `total`:

> <span style="color:#43c6ac">We can place comments inside the code using the hash symbol `#`.</span>

````python
total = 910 + 825 + 1070 + 940

# Show the result
print(total)
````

We can also calculate the average weekly concentrate production and assign it to a variable called `mean`:

````python
# Number of weeks is 4
mean = total / 4

# Show the result
print(mean)
````

To calculate the profit produced by the mine, we need to consider that the price of concentrate per ton is 5 dollars and the operaton cost for that month was 10,000 dollars.

We will also use the following formula:

<center> 
    $ \Large \text{profit} = total_{concentrate} \times price_{concentrate} - cost_{operation} $ 
</center> 
<br>

We can assign the price of concentrate to the variable `concentrate_price` and the operation cost to the variable `operation_cost`:

````python
# Define the variables
concentrate_price = 5
operation_cost = 10000
````

````python
# Calculate the profit
profit = (total * concentrate_price) - operation_cost
print(profit)
````

We can also calculate the variance of the production using the following formula:

<center>
    $ \Large s^{2} = \frac{\sum (x - \bar{x})^{2}}{n - 1} $
</center>

Where:

> $s^{2}$ : variance\
> $x$ : data point\
> $\bar{x}$ : mean\
> $n$ : number of data points

We start by assigning `n` as the number of weeks for productions of 910, 825, 1070, and 940 tons:

````python
n = 4
````

Then we calculate the squared difference between each weekâ€™s value and the weekly average:

````python
x1 = (910 - mean) ** 2   # 1st week
x2 = (825 - mean) ** 2   # 2nd week
x3 = (1070 - mean) ** 2  # 3rd week
x4 = (940 - mean) ** 2   # 4th week
````

Now we sum all these values:

````python
sum_x = x1 + x2 + x3 + x4
````

And finally, we calculate the variance:
> There is a quick and simple way to calculate the variance, it will be shown later on the Numpy tutorial.

````python
variance = sum_x / (n - 1)

# Show the result
print(variance)
````

If we want to calculate the standard deviation (square root of the variance), we can assign it to a variable called `sigma`:

````python
sigma = variance ** 0.5

# Show the result
print(sigma)
````

**Question: If we have two variables `a` and `b` with values 3 and 5, how can we swap the values of both variables using a single line of code?**

Answer: by using an expression like `a, b = b, a`.

> We can define multiple variables on a single line by placing each element in order and separated by commas.

We will start by defining the variables:

````python
a, b = 3, 5

# Show the result
print(a)
print(b)
````

In a single line of code, we can swap the values of these two variables:

````python
a, b = b, a

# Show the result
print(a)
print(b)
````

### <span style="color:lightgreen">**What is an f-string?**</span>
***

<span style="color:gold">f-strings</span> are an advanced way to format text strings in Python. They allow us to insert variables and expressions directly into a text string using curly braces and the `f` prefix.

The basic syntax of an f-string is as follows:

```python
f"text {variable} text"
```

This way, <span style="color:#43c6ac">we eliminate the need to concatenate strings</span> using plus signs and conversions with the `str` function.

> The main advantage of this approach is that it makes our code more readable and efficient.

***
**Example: Using f-strings in sentences**

To turn a sentence into a dynamic string with f-strings, we must insert a placeholder for each variable using curly braces `{}`.

The variables we will use are `name` and `texture`, and we will create an f-string called `phrase`:

````python
# Define the variables
name = "granite"
texture = "phaneritic"
phrase = f"A {name} is a rock of {texture} texture, it contains quartz and feldspar."

# Show the result
print(phrase)
````

***
**Example: RQD calculation**

We are going to calculate the RQD in a core section and display the result using an f-string (as a percentage with 2 decimal places of precision). The length of each core fragment (in cm) is: $8, 12, 15, 9, 25, 19, 11, 4$.

> <span style="color:gold"> Rock Quality Designation (RQD)</span> is a geotechnical index that measures the percentage of intact core pieces longer than 10 cm recovered from a drill run, indicating the quality and degree of fracturing of the rock mass.

We will start by summing the lengths of the fragments greater than 10 cm and dividing by the total length.

````python
# Sum the lengths greater than 10 cm and divide by the total length
rqd = (12 + 15 + 25 + 19 + 11) / (8 + 12 + 15 + 9 + 25 + 19 + 11 + 4)

# Show the result
print(f"RQD: {rqd}")
````

To modify the format of this result, we use the following expression:

```python
f"text: {variable:.3f}"
```

This f-string will transform the numeric value into a numeric value with 3 decimal places of precision.

````python
# Round to 3 decimal places
print(f"RQD: {rqd:.3f}")
````

To modify the format of this result into a percentage, we use the following expression:

```python
f"text: {variable:.2%}"
```

This f-string will transform the numeric value into a percentage value with 2 decimal places of precision.

````python
# Convert to percentage with 2 decimal places
print(f"RQD: {rqd:.2%}")
````

***
**Question: Can we round a number without using f-strings?**

Answer: Yes, for this we use the `round` function:

> The `round` function takes two parameters:
> - The numeric value to be rounded, and
> - The number of decimal places of precision for the result.

We can specify the number of decimal places right after the value we want to round:

````python
# Round pi to 3 decimal places
round(3.141592, 3)
````

### <span style="color:lightgreen">**The `input` function**</span>
***

To finish, letâ€™s look at a brief example of the `input` function. This function allows the user to enter data from the keyboard,
<span style="color:#43c6ac">and it returns the entered data as a text string (str).</span>

The basic syntax of the `input` function is:

```python
variable = input("Message to the user: ")
```

***
**Example: Name input**

We can use the `input` function to ask the user to enter their name and then greet them:

````python
# Enter the data
name = input("Please type your name: ")

# Show the result
print(f"Hello {name}!!!")
````

***
**Example: Rock name input**

We can use the `input` function to ask the user to enter the name of a rock.

We will assign the result to the variable `rock` and display it with `print`:

````python
# Enter the data
rock = input("Enter the name of a rock: ")

# Show the result
print(rock)
````

***
**Example: Well depth**

We can ask the user to enter an integer representing the depth of a well.

To do this, we will convert the result of the `input` function to an integer using the `int` function.

````python
# Enter the data
depth = int(input("Enter the depth of the well in meters: "))

# Show the result
print(f"The depth is {depth} meters.")
````

***
**Example: Total depth of wells**

We can ask the user to enter the depths of two wells and calculate the total depth:

To do this, we will convert the result of the `input` function to a floating-point number using the `float` function.

````python
# Enter the data
depth1 = float(input("Enter the depth of well 1 in meters: "))
depth2 = float(input("Enter the depth of well 2 in meters: "))

# Show the result
print(f"The total depth is {depth1 + depth2:.0f} meters.")
````
