### <span style="color:lightgreen">**What is computational logic?**</span>
***

Computational logic represents a set of techniques for solving problems through the creation of **algorithms**. 

> An <span style="color:gold">algorithm</span> is a set of finite and precise steps used to solve a problem or perform a specific task.

In Python, we can use control flow structures to create algorithms and make logical decisions in our code.

<span style="color:#43c6ac">In general terms, algorithms are a series of instructions that must be followed in a specific order to achieve a goal.</span>

Algorithms can be expressed in different forms, such as pseudocode, flowcharts, programming languages, and other formats.
It is important to note that algorithms must be precise, efficient, and able to solve the problem for which they were designed.

### <span style="color:lightgreen">**Comparisons between values**</span>
***

Yes, it is possible to compare objects in Python.

In Python, comparison operators are used to compare two values and determine whether they are equal or if one is greater or less than the other. These operators are commonly used in control flow statements, such as conditionals and loops.

If we take two numbers as an example, we can perform the following comparisons:

- ‚ÄúGreater than‚Äù `>` and ‚Äúless than‚Äù `<` operations:

````python
# Is 2 plus 2 greater than 5?
(2 + 2) > 5
````

````python
# Is 3 squared less than 10?
(3 ** 2) < 10
````

- ‚ÄúGreater than or equal to‚Äù `>=` and ‚Äúless than or equal to‚Äù `<=` operations:

````python
# Is 3 times 5 greater than or equal to 15?
(3 * 5) >= 15
````

````python
# Is 20 divided by 2 less than or equal to 5?
(20 / 2) <= 5
````

- And finally, ‚Äúequal to‚Äù `==` and ‚Äúnot equal to‚Äù `!=` operations:

````python
# Is 4 times 4 equal to 16?
(4 * 4) == 16
````

````python
# Is 5 times 5 different from 25?
(5 * 5) != 25
````

As you can see, comparisons return <span style="color:gold">boolean values</span>, which can only be `True` or `False`.

### <span style="color:lightgreen">**Logical operators**</span>
***

In Python, the logical operators `or`, `and`, `not` and `in` are used to perform logical operations on boolean expressions.

- `or` returns `True` if at least one of the boolean expressions is `True`. <br>
If all expressions are `False`, it returns `False`.

- `and` returns `True` only if all the boolean expressions are `True`. <br>
If at least one expression is `False`, it returns `False`.

- `not` returns the opposite value of a boolean expression. <br>
If the expression is `True`, it returns `False`; if the expression is `False`, it returns `True`.

- `in` checks whether an object belongs to another (for example, in sequences such as strings or lists).<br>
If the object is contained in the other, it is `True`; otherwise, it is `False`.

````python
True or False
````

````python
False or False
````

````python
True and True
````

````python
True and False
````

````python
not False
````

````python
# Is there a letter "a" in "sandstone"?
"a" in "sandstone"
````

<span style="color:#43c6ac">We can chain comparisons using logical operators, resulting in a boolean value.</span>

***
**Example:** Evaluate the following logical expressions:

`2 is greater than 5, and 3 is less than 10`

````python
(2 > 5) and (3 < 10)
````

`15 squared is equal to 225, or 2 to the eighth power is 250`

````python
(15 ** 2 == 225) or (2 ** 8 == 250)
````

````python
(20 % 2 == 0) and (20 > 10) and (20 ** 2 > 100)
````

***
The `any` function checks if at least one of the expressions inside is `True`:

> Equivalent to using `or` between expressions.

````python
any([(2 > 5), (3 ** 2 == 9), (10 > 2 ** 4)])
````

The `all` function checks if all of the expressions are `True`:

> Equivalent to using `and` between expressions.

````python
all([(2 > 5), (3 ** 2 == 9), (10 > 2 ** 4)])
````

### <span style="color:lightgreen">**Conditionals**</span>
***

Conditionals are control structures that allow a program to take different paths based on certain conditions.

In Python, the reserved words `if`, `else` and `elif` are used to build these structures.

> <span style="color:#43c6ac">In Geology, we can use conditionals to model different scenarios and perform various analyses.</span>

The general syntax of a conditional is as follows:

```python
if condition:
    # Code to execute if the condition is true
elif condition:
    # Code to execute if the previous condition is false
else:
    # Code to execute if all the previous conditions are false
```

<span style="color:#43c6ac">It is important to indent the code belonging to a conditional, using four spaces or the `Tab` key.</span>

***
**Example: Mineral identification**

Suppose we have data for a mineral and we want to identify it based on its physical properties.

In this example, we will use conditionals to evaluate the hardness and color of a mineral in order to identify it:

````python
# Mineral data
hardness = 4
color = "green"
````

We will implement criteria to identify 3 minerals: quartz, malachite, and calcite.

- If the mineral‚Äôs hardness is equal to 7 and its color is transparent, then the mineral is quartz.
- If the mineral‚Äôs hardness is equal to 3 and its color is white, then the mineral is calcite.
- If the mineral‚Äôs hardness is equal to 4 and its color is green, then the mineral is malachite.
- Finally, if the mineral does not meet the identification criteria, we get the phrase "The mineral could not be identified."

````python
# Identify the mineral based on its characteristics
if (hardness == 7) and (color == "transparent"):
    print("The mineral is quartz.")
elif (hardness == 3) and (color == "white"):
    print("The mineral is calcite.")
elif (hardness == 4) and (color == "green"):
    print("The mineral is malachite.")
else:
    print("The mineral could not be identified.")
````

***
**Example: Rock classification**

Suppose we have data from a rock sample and we want to classify it into one of three possible categories: igneous, sedimentary, or metamorphic. To do this, we can use a conditional in Python to evaluate the values of certain chemical elements that are characteristic of each rock type.

In this example, we will use the values of `SiO2` and `MgO` to classify the sample:

````python
# Rock sample data
SiO2 = 75
MgO = 2.5

# Classification thresholds
threshold_igneous = 65  # SiO2
threshold_sedimentary = 5  # MgO
threshold_metamorphic = 5  # SiO2

# Evaluate SiO2 and MgO values to classify the rock
if SiO2 >= threshold_igneous and MgO < threshold_sedimentary:
    print("The rock sample is igneous.")
elif SiO2 < threshold_metamorphic and MgO >= threshold_sedimentary:
    print("The rock sample is sedimentary.")
else:
    print("The rock sample is metamorphic.")
````

In this example, classification thresholds were set based on the typical values of `SiO2` and `MgO` for each rock type.

Then, through a conditional, the values of `SiO2` and `MgO` in the sample were evaluated to determine its type:

- If the `SiO2` values are high and the `MgO` values are low, the rock is classified as igneous.
- If the `SiO2` values are low and the `MgO` values are high, the rock is classified as sedimentary.
- If the values of both elements fall within the threshold corresponding to metamorphic classification, then it is classified as metamorphic.

***
**Example: Geochemical comparison**

Suppose we have two samples and we want to compare them geochemically to determine if they have a similar composition.

In this example, we will use a comparison threshold of 0.5 units and the `abs` function to get the absolute value of the difference between the iron concentrations in the two samples.

````python
# Sample data
sample_1, sample_2 = 3.7, 4.1

# Set a comparison threshold
threshold = 0.5
````

We will use a conditional to compare the difference between both samples with the threshold:

- If the difference is less than or equal to the threshold, a message will be displayed indicating that the concentrations are similar.
- If the difference is greater than the threshold, a message will be displayed indicating that the concentrations are different.

````python
# Compare the iron concentrations in the two samples
if abs(sample_1 - sample_2) <= threshold:
    print("The concentrations in the two samples are similar.")
else:
    print("The concentrations in the two samples are different.")
````

This is just a simple example of how you could use conditionals in Python to compare geological data.
