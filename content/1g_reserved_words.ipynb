### <span style="color:lightgreen">What are reserved words?</span>
***

<span style="color:gold">Reserved words</span> are predefined terms in Python that have a special meaning.

These terms are fundamental to the structure and operation of the code, as they represent commands or instructions that the Python interpreter recognizes and executes in a specific way.

Due to their importance, these words cannot be used as identifiers, which means you cannot use them to name variables, functions, classes, or any other element of your code. This helps prevent confusion and errors, ensuring that the code is clear and understandable both for the interpreter and for developers.

Each reserved word fulfills a specific function within the language:

> **Variables:**
> - `None`: constant that represents the absence of a value.
> - `del`: deletes a variable, list, or any other object.

> **Computational logic:**
> - `True`: boolean constant representing true.
> - `False`: boolean constant representing false.
> - `and` (`&`): logical operator that returns true if both expressions are true.
> - `or` (`|`): logical operator that returns true if at least one expression is true.
> - `not` (`~`): logical negation operator.
> - `if`: declares a conditional structure.
> - `elif`: part of a conditional structure, similar to â€œelse ifâ€ in other languages.
> - `else`: part of a conditional structure, executes the block if none of the previous conditions are true.
> - `is`: identity operator, checks whether two references point to the same object.

> **Definite and indefinite loops:**
> - `for`: declares a loop that iterates over a sequence.
> - `in`: operator used to check whether a value is in a sequence.
> - `while`: declares a loop that runs while a condition is true.
> - `pass`: indicates a null operation; used as a placeholder.
> - `continue`: skips the current iteration of the loop and continues with the next one.
> - `break`: terminates the current loop.

> **Functions and classes:**
> - `def`: declares a new function or method.
> - `assert`: used for assertion testing; raises an exception if the condition is false.
> - `global`: declares that a variable inside a function is global.
> - `nonlocal`: declares that a variable inside a nested function refers to a variable in the enclosing scope.
> - `return`: exits a function and optionally returns a value.
> - `lambda`: creates an anonymous (lambda) function.
> - `yield`: pauses a function and returns a value to the caller; used in generators.
> - `class`: declares a new class.

> **Resource management and exceptions**
> - `with`: simplifies resource management such as file handling.
> - `try`: attempts to execute a block of code and handles exceptions.
> - `except`: handles exceptions in a try/except structure.
> - `finally`: block that is always executed at the end of a try/except structure.
> - `raise`: raises an exception.

> **Library imports:**
> - `import`: imports modules.
> - `as`: used to create an alias when importing modules.
> - `from`: used to import specific parts of a module.

> **Asynchronous programming:**
> - `async`: declares an asynchronous function or context.
> - `await`: waits for the completion of an asynchronous call.


The correct understanding and use of these words is crucial for writing Python programs that are efficient, effective, and free of syntactic errors.

### <span style="color:lightgreen">**Examples**</span>
***

Letâ€™s review the use of some of these reserved words through basic examples:

***
**`None` example:**

```python
# Create a variable x
x = None
print(x)  # Display the value of x (nothing!!!)
```

***
**`del` example:**

```python
# Create a variable x
x = 0
print(f"x is equal to {x}")  # Display the value of x

del x  # Delete the variable x
print(x)  # This causes an error
```

***
**`True`, `False`, `if`, `elif`, `else`, `and`, `or`, `not` example:**

```python
# Conditionals
if (2 + 2 == 5) and False:  # First condition
    print("2 + 2 is equal to 5 and 3 + 3 is equal to 6")
elif (4 + 4 == 9) or (not (1 == 1)):  # Second condition
    print("4 + 4 is equal to 9 or 1 is not equal to 1")
elif True:  # Third condition
    print("The true condition was met")
else:  # If none of the conditions are met
    print("No condition was met")
```

***
**`for`, `in` example:**

```python
# Definite (counted) loop
for number in range(0, 5):
    print(number)
```

***
**`while` `pass`, `continue`, `break` example:**

```python
# Indefinite loop (without using "break")
x = 0

while x < 5:
    print(x) # Show the value of x
    x += 1   # Update the value of x by adding 1
```

```python
# Indefinite loop (using "break")
x = 0

while True:
    if x < 5:
        pass  # Does nothing in this part
    elif 5 < x < 9:
        x += 1
        continue  # Starts the next iteration
    elif x == 10:
        print(f"x is equal to {x}")
        break

    print(f"x is equal to {x}")  # Displays the value of x
    x += 1  # Updates the value of x by adding 1

print("The loop has ended")
```

***
**`def`, `assert`, `return` example:**

```python
def summatory(n):
    """Sum of the first n natural numbers."""

    # Validation
    assert n > 0, "n must be positive"
    assert type(n) == int, "n must be an integer"
    
    result = int(n * (n + 1) / 2)
    return result

# Code test
summatory(10)
```

***
**`lambda` example:**

```python
# Sum of the first n natural numbers
summatory = lambda n: int(n * (n + 1) / 2) if ((n > 0) and (type(n) == int)) else print("Error!")

# Code test
summatory(10)
```

***
**`global`, `nonlocal` example:**

```python
x = 10

def counter():
    num = 0  # Variable in the enclosing scope
    
    def increase():  # Nested function
        nonlocal num  # Indicates we will use the variable num from the enclosing scope
        global x  # Indicates we will use the global variable x
        print(f"The global variable x is equal to {x}")  # Displays the value of x
        num += 1  # Updates num by adding 1
        return x + num  # Returns the sum of x and num
        
    return increase

# Create an instance of the function 'incrementar'
increase_counter = counter()

# Code test
print(increase_counter())  # Should print 11
print(increase_counter())  # Should print 12
print(increase_counter())  # Should print 13
```

***
**`yield` example:**

```python
def counter_until(n):
    counter = 1  # Starts at 1
    while counter <= n:
        yield counter
        counter += 1

# Code test
for number in counter_until(5):
    print(number)
```

***
**`with` example:**

```python
# Create a file and write two lines of text
with open("example.txt", "w") as file:  # w is for write
    # Write two lines to the text file
    file.write("Hello, world!\n")  # First line
    file.write("This is an example of using 'with' in Python.")  # Second line

# Read the file to verify its content
with open("example.txt", "r") as file:  # r is for read
    content = file.read()
    print(content)
```

***
**`try`, `except`, `finally` example:**

```python
def divide(a, b):
    # By adding a try/except/finally structure,
    # we ensure that the division does not allow
    # zero to be used as the denominator
    try:  # This executes the main code of the function
        result = a / b
    except ZeroDivisionError:  # This runs if the error occurs
        print("Error: Division by zero is not allowed.")
        return None
    finally:  # This always executes
        print("The division operation has finished.")
        
    return result

# Code test
print(divide(10, 2))  # Normal case
print("")
print(divide(10, 0))  # Case where the error occurs
```

***
**Ejemplo de `raise`:**

```python
print(2 + 2)
raise AssertionError  # Raises an error
print(3 + 3)  # This code is never executed
```
